```python
#Primera pregunta 
# 1.Que genero es mas rechazado, por que ?
# Calcular la tasa de rechazo para cada género
rechazo_mujeres = dating_real_dataframe[mujeres]['match'].value_counts(normalize=True)[0]
rechazo_hombres = dating_real_dataframe[hombres]['match'].value_counts(normalize=True)[0]

rechazo_mujeres, rechazo_hombres
#Tasa de rechazo para mujeres: 88.32% 
#Tasa de rechazo para hombres: 81.59%
```

```python
# 1.Que genero es mas aceptado, por que ?
# Calcular la tasa de aceptacion para cada género
aceptacion_mujeres = dating_real_dataframe[mujeres]['match'].value_counts(normalize=True)[1]
aceptacion_hombres = dating_real_dataframe[hombres]['match'].value_counts(normalize=True)[1]

aceptacion_mujeres, aceptacion_hombres
#Tasa de rechazo para mujeres: 11.67% 
#Tasa de rechazo para hombres: 18.41%
```

```python
# Que caracteristicas tienen en comun los usuarios rechazados
# Características de las mujeres rechazadas
caracteristicas_mujeres_rechazadas = dating_real_dataframe[mujeres & (dating_real_dataframe['match'] == 0)][['age','gender','Attractive','Intelligent','Fun','Ambitious']].mode() 
salario_mujeres_rechazadas = dating_real_dataframe[mujeres & (dating_real_dataframe['match'] == 0)]['Income'].mean() 


# Características de los hombres rechazados
caracteristicas_hombres_rechazados = dating_real_dataframe[hombres & (dating_real_dataframe['match'] == 0)][['age','career','Attractive','Intelligent','Fun','Ambitious']].mode()
salario_hombres_rechazados = dating_real_dataframe[hombres & (dating_real_dataframe['match'] == 0)]['Income'].mean()


caracteristicas_mujeres_rechazadas.iloc[0],salario_mujeres_rechazadas, caracteristicas_hombres_rechazados.iloc[0],salario_hombres_rechazados
```

```python



# Calcular la tasa de "likes" para cada edad
likes_por_edad = dating_real_dataframe.groupby('age')['match'].mean()

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.plot(likes_por_edad.index, likes_por_edad.values, marker='o')
plt.xlabel('Edad')
plt.ylabel('Probabilidad de recibir un "like"')
plt.title('Probabilidad de recibir un "like" según la edad')
plt.grid(True)
plt.show()

# Encontrar la edad con la mayor tasa de "likes"
edad_optima = likes_por_edad.idxmax()
tasa_maxima_likes = likes_por_edad.max()

print(f"La edad óptima para recibir más 'likes' es {edad_optima} con una tasa de 'likes' de {tasa_maxima_likes:.2f}")
```

```python
 # tiene correlacion el sueldo con la probabilidad de aceptacion ?
 relacion_ingresos_probabilidad = dating_real_dataframe[['Income', 'prob']].corr()
 relacion_ingresos_probabilidad
 #Coeficiente de correlación (0.57): Indica una correlación positiva moderada entre los ingresos y la probabilidad de ser aceptado. Esto sugiere que, en general, a medida que los ingresos aumentan, también tiende a aumentar la probabilidad de ser aceptado.
 
```

```python
plt.scatter(dating_real_dataframe['Income'], dating_real_dataframe['prob'])
plt.xlabel('Ingresos')
plt.ylabel('Probabilidad de ser aceptado')
plt.title('Correlación entre Ingresos y Probabilidad de ser Aceptado')
plt.show()
```

```python




# Filtrar mujeres inteligentes y guapas
mujeres_inteligentes_guapas = dating_real_dataframe[(dating_real_dataframe['gender'] == 0) & 
(dating_real_dataframe['Intelligent'] >= 8) & (dating_real_dataframe['Attractive'] >= 8)]

# Filtrar hombres inteligentes y guapos
hombres_inteligentes_guapos = dating_real_dataframe[(dating_real_dataframe['gender'] == 1) & (dating_real_dataframe['Intelligent'] >= 8) & (dating_real_dataframe['Attractive'] >= 8)]

# Calcular la tasa de aceptación para mujeres inteligentes y guapas
probabilidad_aceptacion_mujeres_inteligentes_guapas = mujeres_inteligentes_guapas['match'].mean()

# Calcular la tasa de aceptación para hombres inteligentes y guapos
probabilidad_aceptacion_hombres_inteligentes_guapos = hombres_inteligentes_guapos['match'].mean()

probabilidad_aceptacion_mujeres_inteligentes_guapas, probabilidad_aceptacion_hombres_inteligentes_guapos
# 14.28% de ser aceptadas
# 50% de ser aceptados
```

```python
# Datos para la visualización
categorias = ['Mujeres Inteligentes y Guapas', 'Hombres Inteligentes y Guapos']
probabilidades = [probabilidad_aceptacion_mujeres_inteligentes_guapas, probabilidad_aceptacion_hombres_inteligentes_guapos]

# Crear el gráfico de barras
plt.bar(categorias, probabilidades, color=['blue', 'green'])
plt.xlabel('Categoría')
plt.ylabel('Probabilidad de Aceptación')
plt.title('Probabilidad de Aceptación por Categoría')
plt.ylim(0, 1)
plt.show()
```

```python
#¿Cuáles son las actividades más populares entre los hombres y las mujeres?
actividades_columnas = dating_real_dataframe.loc[:, 'sports':'yoga']

actividades_hombres_promedio = dating_real_dataframe[hombres].loc[:, 'sports':'yoga'].mean()
actividades_mujeres_promedio = dating_real_dataframe[mujeres].loc[:, 'sports':'yoga'].mean()

actividades_promedio = pd.DataFrame({'Hombres': actividades_hombres_promedio,'Mujeres': actividades_mujeres_promedio})

actividades_hombres_promedio, actividades_mujeres_promedio
```
